````python
def generateAccountSnapshots(group):
    group = group.sort_values('created_at')
    print("------------------------------ ")

    min_date = group['created_at'].min()
    account_age_days = (group['created_at'] - min_date).dt.days
    group['account_age_days'] = account_age_days
    group['is_currently_in_arrears'] = group['account_status'].apply(lambda x: True if x == 'Arrears' else False)

    return group
````

````python
fact_accounts_snapshot_df = fact_accounts_snapshot_df.groupby("id").apply(generateAccountSnapshots)
````

````python
from uuid import UUID

# ── Clean dataframe first ─────────────────────────────────────────────────────
string_cols = ['lead_id', 'customer_id', 'agent_id']
for col in string_cols:
    facts_leads_funnel_df[col] = facts_leads_funnel_df[col].fillna('').astype(str)

# UUID column — must stay as string in correct UUID format
facts_leads_funnel_df['lead_source_id'] = facts_leads_funnel_df['lead_source_id'].astype(str)

# Boolean → int
bool_cols = ['converted_to_account', 'converted_to_sale', 'converted_to_installation']
for col in bool_cols:
    facts_leads_funnel_df[col] = facts_leads_funnel_df[col].astype(int)

# Days sentinel → None
days_cols = ['days_to_account', 'days_to_sale', 'days_to_dispatch']
for col in days_cols:
    facts_leads_funnel_df[col] = facts_leads_funnel_df[col].replace(99999, None)
    facts_leads_funnel_df[col] = facts_leads_funnel_df[col].where(
        facts_leads_funnel_df[col].notna(), None
    )

# ── Build records ─────────────────────────────────────────────────────────────
records = []
for _, row in facts_leads_funnel_df.iterrows():
    # Validate UUID format before inserting
    try:
        uuid_val = str(UUID(str(row['lead_source_id'])))
    except (ValueError, AttributeError):
        print(f"Skipping row with bad lead_source_id: {row['lead_source_id']}")
        continue

    records.append((
        str(row['lead_id']),
        str(row['customer_id']),
        uuid_val,                                    # validated UUID string
        str(row['agent_id']) if row['agent_id'] else None,
        row['created_at'],
        int(row['converted_to_account']),
        int(row['converted_to_sale']),
        int(row['converted_to_installation']),
        row['days_to_account'],                      # None or int
        row['days_to_sale'],
        row['days_to_dispatch'],
        float(row.get('lead_quality_score', 0.0)),
        # created_at has DEFAULT now() so omit it
    ))

print(f"Records to insert: {len(records)}")

# ── Bulk insert ───────────────────────────────────────────────────────────────
client.execute(
    """
    INSERT INTO sunculture_db.fact_lead_funnel
    (lead_id, customer_id, lead_source_id, agent_id, lead_created_date,
     converted_to_account, converted_to_sale, converted_to_installation,
     days_to_account, days_to_sale, days_to_dispatch, lead_quality_score)
    VALUES
    """,
    records
)
print(f"✓ Inserted {len(records)} rows successfully")
````
